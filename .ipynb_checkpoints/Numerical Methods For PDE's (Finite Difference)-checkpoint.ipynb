**Q2.** Write a program to compute $15!$(factorial).

Solution:

```python
n = 15
factorial = 1
for i in range(1,n+1):
    factorial = factorial * (i)
    
print(f"{n}!= {factorial}")

# Alternatively using Recursion.
def factorial(n):
   if n == 1:
       return n
   else:
       return n*factorial(n-1)
    
print(f"{n}!= {factorial(n)}")
```

**Q4.** Write a function that solves a quadratic equation $$ax^2+bx+c=0$$

Solution:

```python
import cmath

def Quadratic_Solver(a,b,c):
    discriminant = b**2 - 4*a*c
    d = cmath.sqrt(discriminant)
    x1 = (-b+d)/(2*a)
    x2 = (-b-d)/(2*a)
    return x1, x2

def main():
    x1, x2 = Quadratic_Solver(1,2,2)
    print(f'First root is: {x1}')
    print(f'Second root is: {x2}')

if __name__ == "__main__":
    main()
```

**Q8.** Write a program that uses Newton-Rapshon method to calculate the roots of a polynomial function $f(x) = x^3-23x^2+142x-120$. Note the iterative scheme for N-R Method is $x_{n+1}=x_{n}-\frac{f(x_n)}{f^{'}(x_n)}$.

Solution:

```python
def f_xn(x):
    f_xn = x**3 - 23*x**2 + 142*x - 120 
    return f_xn

def fprime_xn(x):
    fprime_xn = 3*x**2 - 46*x + 142 
    return fprime_xn

def newton(xn):
    difference = 1
    tolerance = 1.0e-8
    xnp1 = 0 
    while difference > tolerance:
        xnp1 = xn - (f_xn(xn)/fprime_xn(xn))
        difference = np.abs(xnp1-xn)
        xn = xnp1
    print(f'One of the root of this function is {xn}.')
    
xn = float(input('Enter Guess value for the root of the function: '))

newton(xn)
```

**Q11.** Use Archimedes Method of approximating $\pi$ using polygons. Note $\pi = \frac{C}{D}$, so taking $D = 1$, we get $C = \pi$. Therefore, We can use perimeter of Polygon to approximate $\pi$. The approximation formula Archimedes came up with is
$$P = \lim_{n\to\infty} \frac{n}{\sqrt{2}}\sqrt{1 - \cos (\frac{360}{n})} = \pi$$
Also use Area of polygon to approximate $\pi$. This time using $A = \pi r^2$ with $r=1$, we get $A=\pi$. Thus, area of polygon equals $\pi$. The approximation formula is 
$$A = \lim_{n\to\infty} \frac{n}{2}\sin{(\frac{360}{n})} = \pi$$

Solution:

```python
import numpy as np

N = 100000    # Number of sides of Polygon.

# Approximating pi using Perimeter of Polygon.
for n in np.arange(3,N+1):
    P = (n/np.sqrt(2))*(np.sqrt(1-np.cos(np.deg2rad(360) / n)))
pi = P
print(f"Approximate value of π using {N}-sided polygon perimeter is:", "{:.10f}".format(pi))

# Approximating pi using Area of Polygon.
for n in np.arange(3,N+1):
    A = (n/2)*np.sin(np.deg2rad(360)/n)
pi = A
print(f"Approximate value of π using {N}-sided polygon perimeter is:", "{:.10f}".format(pi))
```

# 2 Finite Difference Methods

Finite Difference Methods approximates the Partial Differential Equation(PDE) using Taylor series at discrete points in the domain. This creates a system of Algebraic equations which can be solved to find the solution of PDE. The solution to the algebraic equation is an approximation of the solution of the PDE. The accuracy of the approximate solution depends upon the order of the FD Method used and the size of step length. The smaller the step size gives better accuracy but at a higher computational cost.

## The Taylor Series
The Taylor Series is given as:
$$
\begin{aligned}
 F(x+h) &= \sum_{n=0}^\infty \frac{h^n}{n!} F_{(n)}(x) \\
        &= F(x) + hF_{(x)}(x) + \frac{h^2}{2}F_{(xx)}(x) + \frac{h^3}{6}F_{(xxx)}(x) + \dots
\end{aligned}
$$
where $h$ is the step size and $F$ is a function of $x$.

**Backward Taylor** is given by replacing $h$ by $-h$.
$$
\begin{aligned}
 F(x-h) &= \sum_{n=0}^\infty (-1)^n \frac{h^n}{n!} F_{(n)}(x) \\
        &= F(x) - hF_{(x)}(x) + \frac{h^2}{2}F_{(xx)}(x) - \frac{h^3}{6}F_{(xxx)}(x) + \dots
\end{aligned}
$$

## Truncating Taylor Series
The **truncation error** is occurred when we omit the higher order terms in the Taylor series. The truncation error is expressed as $\mathcal{O}(h^n)$ which means $\mathcal{O}(h^n) \rightarrow 0$ as $h \rightarrow 0$ is proportional to $h^n$. If the truncation error is $\mathcal{O}(h^2)$ then halving $h$ will reduce truncation error by $(\frac{1}{2})^2 = \frac{1}{4}$.

## 2.1 First order & Second order Difference

### First Order Difference
* **First Order Forward Difference:**
Truncating Taylor series and rearranging $F_x(x)$ as subject, we obtain first order forward difference approximation of $F_x(x)$.
\begin{align}
F(x+h) = F(x) + hF_x(x) + \mathcal{O}(h^2)\\
 \boxed{F_x(x) = \frac{F(x+h)-F(x)}{h} + \mathcal{O}(h)} 
\end{align}
* **First Order Backward Difference:**
\begin{align}
F(x-h) = F(x) - hF_x(x) + \mathcal{O}(h^2)\\
\boxed{F_x(x) = \frac{F(x)-F(x-h)}{h} + \mathcal{O}(h)} 
\end{align}
This is the First Order Backward approximation of $F_x(x)$.

### Second Order Central Difference
* To obtain Second Order Central Difference approximation of $F_x(x)$, we truncate Forward and Backward Taylor series to second order and take the difference and rearrange:
\begin{align}
F(x+h) = F(x) + hF_x(x) + \frac{h^2}{2}F_{xx}(x) + \mathcal{O}(h^3) \tag{i}\\
F(x-h) = F(x) - hF_x(x) + \frac{h^2}{2}F_{xx}(x) + \mathcal{O}(h^3) \tag{ii}
\end{align}
$(i)-(ii)$ gives:
\begin{align}
F(x+h)-F(x-h) = 2hF_x(x) + \mathcal{O}(h^3)\\
\boxed{F_x(x) = \frac{F(x+h)-F(x-h)}{2h} + \mathcal{O}(h^2)} 
\end{align}
The Second Order Central Difference is more accurate than the first order difference.

## 2.2 Deriving Higher Order Differences
There are two methods for deriving higher order finite difference approximations:
1. Using Lower order differences.
2. Method of Undetermined Coefficients.(Better Method)

### Using Lower Order Differences
* We can derive $F_{xx}(x)$ using lower-order differences.
\begin{gather}
F_{xx}(x) = \frac{\partial^2 F}{\partial x^2} = \frac{\partial}{\partial x}(\frac{\partial F}{\partial x})\\
\implies F_{xx}(x) = \overbrace{\frac{F_x(x+h)-F_x(x)}{h}}^\text{Forward Difference}\\
= \frac{1}{h}\bigg[\underbrace{\frac{F(x+h)-F(x)}{h}}_\text{backward difference} - \underbrace{\frac{F(x)-F(x-h)}{h}}_\text{backward difference}\bigg]\\
\boxed{F_{xx}(x) = \frac{F(x+h)-2F(x)+F(x-h)}{h^2}}
\end{gather}
This is **Second Order Symmetric difference** approximation of $F_{xx}(x)$.
* $F_{xxx}(x)$ can be approximated by applying Central Difference approximation of $F_x$ to two central difference approximation of $F_{xx}(x)$.
\begin{gather}
F_{xxx}(x) = \frac{\partial}{\partial x}(\frac{\partial^2 F}{\partial x^2})\\
= \frac{1}{2h}[F_{xx}(x+h)-F_{xx}(x-h)]\\
= \frac{1}{2h}\bigg[\frac{F(x+2h)-2F(x+h)+F(x)}{h^2} - \frac{F(x)-2F(x-h)+F(x-2h)}{h^2}\bigg]
\end{gather}
\begin{gather}
\implies \boxed{F_{xxx}(x) = \frac{-F(x-2h)+2F(x-h)-2F(x+h)+F(x+2h)}{2h^3}+\mathcal{O}(h^2)}
\end{gather}
The $n^{th}$Order Finite Difference requires atleast $(2n-1)$ nodes(although some nodes may have zero coefficients).

### Mixed Derivatives
* Mixed derivatives can be derived by taking Central differences for both $x$ and $y$ variables.
\begin{gather}
\frac{\partial^2 F}{\partial x \partial y} = \frac{\partial}{\partial x}\bigg(\frac{\partial F}{\partial y}\bigg)\\
F_{xy} = \frac{1}{2\Delta x}[F_y(x+\Delta x,y)-F_y(x-\Delta x,y)]\\
= \frac{1}{2\Delta x}\bigg[\frac{F(x+\Delta x,y+\Delta y)-F(x+\Delta x,y-\Delta y)}{2\Delta y}-\frac{F(x-\Delta x,y+\Delta y)-F(x-\Delta x,y-\Delta y)}{2\Delta y}\bigg]\\
\boxed{F_{xy} = \frac{1}{4\Delta x \Delta y}[F(x+\Delta x,y+\Delta y)-F(x+\Delta x,y-\Delta y)-F(x-\Delta x,y+\Delta y)+F(x-\Delta x,y-\Delta y)]}
\end{gather}

## 2.3 Method of Undetermined Coefficients

The steps for applying this method goes as follows:
1. Write Finite difference using the neighbouring nodes that you are using multiplied by some coefficient $\alpha_{i}$. For example for 3 points symmetric difference use $F_{n}(x) = \alpha_{1} F(x-h) + \alpha_{2} F(x) + \alpha_{3} F(x+h)$.
2. Replace terms other that $F(x)$ i.e. $F(x+h)$ and $F(x-h)$ with Taylor series expansion to the required order of accuracy. For example if you want $4^{th}$order accuracy then replace with $4^{th}$order Taylor series.
3. Factor out $F$ terms with combination of $\alpha_{i}$'s coefficients..
4. The coefficients of the $F$ terms should all be zero except for derivative you want to approximate.
5. Form a system of equations in terms of $\alpha_{i}$ and solve. Substitute $\alpha_{i}$'s in step 1 formula to get the finite difference approximation.

# 2.4 Analysing Different Order FD Approximations
We will compare finite difference approximations for different order approximations to analyse their accuracy. Since, the derivative of exponential function is itself, we will be approximating $e^x$ using different order of approximation at $x=1$. The exact value of the derivative of exponential function at $x=1$ is $2.7182818\dots$ .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialising variables.
x = 1
forward = np.zeros(5)
backward = np.zeros(5)
central = np.zeros(5)
fourth = np.zeros(5)
err_forward = np.zeros(5)
err_backward = np.zeros(5)
err_central = np.zeros(5)
err_fourth = np.zeros(5)
h = np.zeros(5)    

for i in range(5):
    h[i] = 2**(-(i+1))    # looping over Step size of h=[0.5,0.25,0.125,0.0625,0.03125].
    
    # Calculate finite-difference approximation for 4 different order methods.
    forward[i] = (np.exp(x+h[i]) - np.exp(x))/h[i]
    backward[i] = (np.exp(x) - np.exp(x-h[i]))/h[i]
    central[i] = (np.exp(x+h[i]) - np.exp(x-h[i]))/(2*h[i])
    fourth[i] = (np.exp(x-2*h[i]) - 8*np.exp(x-h[i]) + 8*np.exp(x+h[i]) - np.exp(x+2*h[i]))/(12*h[i])
    
    # Calculate errors in approximation.
    exact = np.exp(1)    # derivative of e^x at 1
    err_forward[i] = np.abs(exact - forward[i])
    err_backward[i] = np.abs(exact - backward[i])
    err_central[i] = np.abs(exact - central[i])
    err_fourth[i] = np.abs(exact - fourth[i])

# Tabulating approximation of different order methods and step sizes.
fd = {"forward": forward, "backward": backward, "central": central, "fourth": fourth}
table1 = pd.DataFrame(fd, index = h) 
table1 = table1.rename_axis("h(step size)", axis="columns")

# Tabulating errors of different order methods and step sizes.
er = {"Error in forward Method": err_forward, "Error in backward Method": err_backward, "Error in central Method": err_central, "Error in fourth Method": err_fourth}
table2 = pd.DataFrame(er, index = h) 
table2 = table2.rename_axis("h(step size)", axis="columns")
```

## Error Analysis
The form of the error in the approximation is of the form $error = C h^n$ where $C$ is some constant, $h$ is step size and $n$ is the order of method. Applying $log$ both sides and dropping $C$ as it does not affect the behaviour, we get $n = \frac{log (error)}{log (h)}$. We can plot the graph of $log (error)$ against $log (h)$ and the gradient of the graph gives the **order of the method**.

```python
# Plotting Log(error) vs Log(h) graph
plt.loglog(h, err_forward, 'b-', label="Forward")
plt.loglog(h, err_backward, 'c-', label="Backward")
plt.loglog(h, err_central, 'm-', label="Central")
plt.loglog(h, err_fourth, 'y-', label="Fourth")
plt.xlabel("$log(h)$")
plt.ylabel("$log(error)$")
plt.title('Log-Log Plot showing Order of Method')
plt.legend(loc='best')
plt.show()
```

```python
# Approximating Order of method using log-log error.
forward_estimation = (np.log(err_forward[0])-np.log(err_forward[4]))/(np.log(h[0]) - np.log(h[4]))  
backward_estimation = (np.log(err_backward[0])-np.log(err_backward[4]))/(np.log(h[0]) - np.log(h[4])) 
central_estimation = (np.log(err_central[0])-np.log(err_central[4]))/(np.log(h[0]) - np.log(h[4])) 
fourth_estimation = (np.log(err_fourth[0])-np.log(err_fourth[4]))/(np.log(h[0]) - np.log(h[4])) 

# Tabulating order of method.
od = {"forward": forward_estimation, "backward": backward_estimation, "central": central_estimation, "fourth": fourth_estimation}
table3 = pd.DataFrame(od,['Order'])
table3 = table3.style.set_caption("Order of Method")
table3
```

Note the forward and backward are approximated as first order method while central is second and fourth is fourth order.

## 2.6 Solving Simple Advection Equation Using FDM

The advection equation $U_t + vU_x = 0$ where $t$ is time and $x\in[a,b]$ is the spatial domain models transport of some material along a channel. The function $U(t,x)$ gives the concentration of some material at time $t$ for point $x$ and $v$ is the advection velocity. If an initial condition of the PDE is given $U(0,x) = g(x)$ then the exact solution of this is $U(t,x) = g(x+vt)$.

<img src="fd7.jpg" width="600" height="300"/>

## Deriving the Finite Difference Scheme for $U_t + vU_x = 0$

We will use Forward Difference to approximate $U_t$ and since $v$ is positive we will use Backward difference to approximate $U_x$. We will get a formula to solve for next value in time given all the previous values. The Forward time-Backward Space is conditionally stable for $0 \leq C = \frac{v\Delta t}{\Delta x} \leq 1$. This is the Courant–Friedrichs–Lewy condition. So, for this question we must have $\Delta t \leq \Delta x$ for stability. A better method would be to use Lax-Friedrich Method which is second-order central in $x$ and forward in $t$. The Forward time-Forward space is unconditionally unstable.

<img src="fd8.jpg" width="600" height="450"/>

The above expression calculates $u_{i}^{n+1}$ and depends upon values of $u_{i}^{n}$ and $u_{i-1}^{n}$. If we take $i=0$ the expression becomes $u_{0}^{n+1} = u_{0}^{n} - \frac{v\Delta t}{\Delta x}(u_{0}^{n} - u_{-1}^{n})$. The $u_{-1}^{n}$ node does not exist and we can only compute solutions for nodes $u_{1}^{n}$ to $u_{N-1}^{n}$ which are called the **computational nodes**. $u_{0}^{n}$ and $u_{N}^{n}$ are called **Boundary nodes** specified by some boundary conditions.

> * Boundary Condition: &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$U(t,0) = 0$,&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$U(t,1) = 0$

\Delta x = \frac{(b-a)}{N} &= \frac{(1-0)}{10} &= 0.1

We will now code the above scheme.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialising Variables
v = 1               # Velocity.
t = 0               # Time.
N = 100             # N = (Number of nodes-1)
a = 0               # Lower bound for spatial domain(x).
b = 1               # Upper bound for spatial domain(x).
dx = (b-a)/N        # Spatial step length.
dt = 0.005          # Time step length.
solution = {}       # Initialising dictionary object to store solutions.

# Initialising Spatial Array
x = np.linspace(a, b, num = N+1)

# Initial Condition
u0 = np.exp(-100*(x-0.4)**2)

# Boundary Conditions
u0[0]=u0[10]=0

u = u0                          # Initialising u.
unp1 = np.zeros(np.shape(u))    # Initialising unp1.

# Calculating 30 time steps of Finite Difference Scheme
for n in np.arange(1,31):
    
    # Looping through computational nodes
    for i in np.arange(1,N):
        unp1[i] = u[i] - v*(dt/dx)*(u[i]-u[i-1])
    
    # Update u and t
    u = unp1
    t = t + dt
    
    # Output Current Solution
#   for i in np.arange(0,N+1):
#       print(u[i])

    # Storing solution in a dictionary object
    solution["u({0})".format(n)] = u.copy()    # must use u.copy() to correctly copy all u(n)'s.

# Plotting solutions.
exact_solution = np.exp(-100*(x - (t*v) - 0.4)**2)
plt.figure(figsize=(9, 5))
plt.plot(x, u0, "y-", label="Initial Conditions")
plt.plot(x, u, "c-", label="FDM Approximation")
plt.plot(x, exact_solution, "r-", label="Exact Solution")
plt.xlabel("$x$")
plt.ylabel("$u$")
plt.title('Plot showing Exact, Initial Condition and FDM Solutions')
plt.legend(loc='best')
plt.show()

# Displaying solution in a table.
table4 = pd.DataFrame.from_dict(solution, orient='index', columns=x)
table4 = table4.rename_axis("x", axis="columns")    
table4  
```

# 3 Higher Order PDE Classification
Second order PDE can be written in standard form as 
\begin{split}
a(x,y)u_{xx} + 2b(x,y)u_{xy} + c(x,y)u_{yy} &= F(x,y,u,u_x,u_y) 
\end{split}
The PDE can be **classified** according to following rule:
> * $b^2-ac > 0 \implies$ Hyperbolic PDE e.g. Wave equation.
> * $b^2-ac < 0 \implies$ Elliptic PDE e.g. Laplace, Poisson equation.
> * $b^2-ac = 0 \implies$ Parabolic PDE e.g. Heat equation.

# 4 Elliptic PDE's
These model Steady state problems meaning solution does not change over time and thus doesn't depend upon time. Hence, the solution are smooth with no discontinuities and the independent variables are all spatial e.g. x and y coordinates for 2-D problem. The solution of Elliptic PDE's doesn't depend either on time or initial condition but does depend on Boundary Conditions. Examples of Elliptic PDE's include:
> * Laplace Equation: $U_{xx} + U_{yy} = 0$ 
> * Poisson Equation: $U_{xx} + U_{yy} = f(x,y)$ 

These equations can be used to model electric/magnetic/gravitational fields.

## 4.1 Discretising a 2-D Domain
We make a rectangular or square grid. If $x\in [a,b]$ and $y\in [c,d]$ with $(N_x +1)$ and $(N_y +1)$ nodes along $x$ and $y$ direction, we have
> * $\Delta x = \frac {b-a}{N_x}$
> * $\Delta y = \frac {d-c}{N_y}$

So, $x_i = a + i \Delta x$ and $y_j = c + j \Delta y$ where $i,j = 0,1,\dots$

[Note: $\Delta x$ & $\Delta y$ are step sizes and $N_x$ and $N_y$ are not derivatives here].
<img src="fd11.jpg" width="450" height="450"/>

## 4.2 Boundary Conditions
<img src="fd9.jpg" width="600" height="200"/>

Depending on the finite difference scheme used it may not be possible to compute values of boundary nodes. The values of boundary nodes are determined by Boundary conditions. There are 3 type of Boundary Conditions:
> * **Dirichlet Boundary Conditions**: The value of function is specified on the boundary i.e. $u(x,y) = f(x,y)$ on $\partial D$ where $\partial D$ is domain boundary.
> * **Neumann Boundary Conditions**: The Normal to the boundary derivative of the function is specified on boundary(given flux condition) i.e. $\frac {\partial u}{\partial n} = f(x,y)$ on $\partial D$.
> * **Robin Boundary Conditions**: A linear combination of the function and its normal derivative is given on the boundary i.e. $a(x,y)u(x,y) + b(x,y)\frac {\partial u(x,y)}{\partial n} = f(x,y)$ on $\partial D$.

## 4.3 Dirichlet Boundary Conditions
Dirichlet Boundary Conditions are used when points on the boundaries are not influenced by the domain i.e. points on boundary are not influenced by whatever going on inside the boundary.
- Boundary values can be constant. 
    - Example:&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_0 = 0$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_N = 0$
- Boundary values depend on some function. 
    - Example:&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_0 = f(x,y)$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_N = g(x,y)$
- Boundary can be used to model periodic behaviour. 
    - Example:&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_0 = u_{N-1}$&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;$u_N = u_1$
    <img src="fd12.jpg" width="600" height="250"/>
    Here, the last computational node is same as the left boundary node and first compulational node is same as the right boundary node.

## 4.4 Neumann Boundary Conditions
These are when the boundary values are influenced by values within the domain. The value of boundary nodes are determined by evaluating derivatives across the boundary. Additional **ghost nodes** are added outside of the domain so that the boundary nodes can be computed using finite difference method.
<img src="fd13.jpg" width="600" height="200"/>

If $u_x = f(u)$ on the Boundary then using Central difference 
\begin{split}
\frac {u_1 - u_{-1}} {2\Delta x} &= f(u)\\
u_{-1} &= u_1 - 2\Delta x f(u)
\end{split}
> * **Transmissive Boundaries**: If $f(u)=0$ i.e. zero gradient across boundary. This is a boundary that allows any flow to escape the domain without resistance. So, $u_{-1} = u_1$, $u_{N+1} = u_{N-1}$.
> * **Solid Boundaries**: No flow through the boundary.e.g. Tank of water. 
If $v$ is velocity then $v_{-1} = -v_{1}$ and $v_{N+1} = -v_{N-1}$ at the boundary.

## 4.5 Solving Laplace Equation with Dirichlet Boundary Condition
\begin{split}
U_{xx} + U_{yy} &= 0
\end{split}
Applying $2^{nd}$order Symmetric difference in both x and y direction, we get
\begin{split}
\frac {u_{i+1,j} - 2u_{i,j} + u_{i-1,j}} {\Delta x^2} + \frac {u_{i,j+1} - 2u_{i,j} + u_{i,j-1}} {\Delta x^2}  &= 0
\end{split}
Setting $\Delta x = \Delta y$ and make $u_{i,j}$ the subject
\begin{split}
u_{i,j} &= \frac {u_{i,j+1} + u_{i+1,j} + u_{i,j-1} + u_{i-1,j}} {4}
\end{split}
The stencil for FDS to solve Laplace equation is: <img src="fd14.jpg" width="250" height="250"/>
The grid for solving laplace is given below where **cross** represents the boundary nodes and **black circle** represents computational nodes we are solving for. For simplicity we have taken 5 nodes in X-direction and 4 nodes in Y-direction.<img src="fd15.jpg" width="450" height="400"/>
The Initial Condition and Boundary Condition for a simple (4 X 5) grid example is given below in blue.<img src="fd16.jpg" width="400" height="400"/>
Note we initialise the computational nodes to zeros but we could equally have chosen any value as we like since Elliptic equation has a steady state solution and its solution is not influenced by the initial condition and we would arrive at the same solution.


Using row-by-column order i.e. starting at first row and sweeping across all columns we get system of equations for all computational nodes.
\begin{split}
u_{1,1} &= \frac {(u_{1,2} + u_{2,1} + 6.8 + 7.2)}{4}\\
u_{2,1} &= \frac {(u_{2,2} + u_{3,1} + u_{1,1} + 7.7)}{4}\\
u_{3,1} &= \frac {(u_{3,2} + u_{2,1} + 9.4 + 8.7)}{4}\\
u_{1,2} &= \frac {(8.9 + u_{2,2} + u_{1,1} + 8.4)}{4}\\
u_{2,2} &= \frac {(8.9 + u_{3,2} + u_{2,1} + u_{1,2})}{4}\\
u_{3,2} &= \frac {(8.9 + 9.2 + u_{3,1} + u_{2,2})}{4}
\end{split}
We can rearrange to write unknown on L.H.S and known values on R.H.S which gives a Linear system. This can be represented in matrix form $A\underline{u} = \underline{b}$ as:
\begin{gather}
\begin{pmatrix} 4 & -1 & 0 & -1 & 0 & 0 \\ -1 & 4 & -1 & 0 & -1 & 0 \\ 0 & -1 & 4 & 0 & 0 & -1 \\ -1 & 0 & 0 & 4 & -1 & 0 \\ 0 & -1 & 0 & -1 & 4 & -1 \\ 0 & 0 & -1 & 0 & -1 & 4 \end{pmatrix} \begin{pmatrix} u_{1,1} \\ u_{2,1} \\ u_{3,1} \\ u_{1,2} \\ u_{2,2} \\ u_{3,2} \end{pmatrix}=\begin{pmatrix} 14 \\ 7.7 \\ 18.1 \\ 17.3 \\ 8.9 \\ 18.1 \end{pmatrix}
\end{gather}
This can easily solved using Gaussian elimination but in practice we have very large matrix and we will use **iterative methods** to solve it.

## 4.5a Jacobi Method (Iterative Method)
This is an indirect method to solve a linear system of equations. The estimates of solution are improved until two successive iterations agree to required accuracy. 
<img src="fd17.jpg" width="350" height="350"/>
The idea is to find $u_{i,j}$ for nodes which are black dots(computational nodes) using iteration. The steps for solving are as follows: 
> * **Step 1**: Take initial guess on distribution of $u_{i,j}$ and call it $u^0_{i,j}$. Now, use \begin{split}
u^1_{i,j} &= \frac {u_{i,j+1} + u_{i+1,j} + u_{i,j-1} + u_{i-1,j}} {4}
\end{split}
to find $u^1_{i,j}$ (i.e. by computing $u_{i,j}$ for all computational nodes without changing boundaries).
> * **Step 2**: Using $u^1_{i,j}$ values for new computational nodes(black dots), iterate the process using \begin{split}
u^{K+1}_{i,j} &= \frac {u^{K}_{i,j+1} + u^{K}_{i+1,j} + u^{K}_{i,j-1} + u^{K}_{i-1,j}} {4}
\end{split}$u^{K}_{i,j}$ denotes the current estimate for $u_{i,j}$ and $u^{K+1}_{i,j}$ denotes the improved estimate. Iteration terminates when difference between successive iterations is very small($\approx 0$).

Now going back to our example and using row-by-column order i.e. starting at first row and sweeping across all columns we get system of equations for all computational nodes for first iteration i.e. $u^1_{i,j}$.
\begin{split}
u^1_{1,1} &= \frac {(0 + 0 + 6.8 + 7.2)}{4} &= 3.5\\
u^1_{2,1} &= \frac {(0 + 0 + 0 + 7.7)}{4} &= 1.925\\
u^1_{3,1} &= \frac {(0 + 0 + 9.4 + 8.7)}{4} &= 4.525\\
u^1_{1,2} &= \frac {(8.9 + 0 + 0 + 8.4)}{4} &= 4.325\\
u^1_{2,2} &= \frac {(8.9 + 0 + 0 + 0)}{4} &= 2.225\\
u^1_{3,2} &= \frac {(8.9 + 9.2 + 0 + 0)}{4} &= 4.525
\end{split}
For the next iteration to get values of $u^2_{i,j}$ we do the same procedure:
\begin{split}
u^2_{1,1} &= \frac {(4.325 + 1.925 + 6.8 + 7.2)}{4} &= 5.0625\\
u^2_{2,1} &= \frac {(2.225 + 4.525 + 3.5 + 7.7)}{4} &= 4.4875\\
u^2_{3,1} &= \frac {(4.525 + 1.925 + 9.4 + 8.7)}{4} &= 6.1375\\
u^2_{1,2} &= \frac {(8.9 + 2.225 + 3.5 + 8.4)}{4} &= 5.7562\\
u^2_{2,2} &= \frac {(8.9 + 4.525 + 1.925 + 4.325)}{4} &= 4.9188\\
u^2_{3,2} &= \frac {(8.9 + 9.2 + 4.525 + 2.225)}{4} &= 6.2125
\end{split}
We continue this until the convergence criteria is satisfied.

## Convergence Criteria
We stop iteration using convergence criteria of Vector Norm($L_2 Norm$)i.e. 
\begin{equation}
\mid \mathbf{x} \mid_2 = \Big(\sum_{i} \mathbf{x^2_i}\Big)^{\frac{1}{2}}
\end{equation}
If $\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid < \mathbf{tolerance}$, we stop iteration i.e.$\mathbf{u}^{k} \approx \mathbf{u}^{k+1} \approx \mathbf{u}$.

Calculating $\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid$ for two iterations we get:
\begin{split}
\mid \mathbf{u}^{1} - \mathbf{u}^{0} \mid &= [(3.5-0)^2 + (1.925-0)^2 + (4.525-0)^2 + (4.325-0)^2 + (2.225-0)^2 + (4.525-0)^2]^\frac{1}{2} = 8.9757\\
\mid \mathbf{u}^{2} - \mathbf{u}^{1} \mid &= [(5.0625-3.5)^2 + (4.4875-1.925)^2 + (6.3175-4.525)^2 + (5.7562-4.325)^2 + (4.9188-2.225)^2 + (6.2125-4.525)^2]^\frac{1}{2}\\
&= 4.8745
\end{split}
We want to continue iterating until the $\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid$ is less than some small prespecified value of **tolerance**.

## Flow Chart for Computation
<img src="fd18.jpg" width="450" height="450"/>

```python
# Code to solve Laplace equation with Dirichlet BC using Jacobi Method.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialising Variables.
Nmax = 4                                 # Nmax = (Number of nodes-1) in X-direction.
Mmax = 3                                 # Mmax = (Number of nodes-1) in Y-direction.
tolerance = 1e-6                         # Tolerance for Convergence Criteria.
error = 1                                # Error.
k = 0                                    # Iteration Counter.
solution = {"0":np.zeros(6)}             # Initialising dictionary object to store solutions.
a = []                                   # Empty list for storing results which will be added to Solution Dictionary.
x = ["$u^k_{1,1}$", "$u^k_{2,1}$", "$u^k_{3,1}$", "$u^k_{1,2}$", "$u^k_{2,2}$", 
     "$u^k_{3,2}$", "$\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid$"]  # Column heading for Solution Table.

# Initialising u Array with Boundary Conditions and Guess values for Computational Nodes (u0).
u = np.array([[8.9, 8.9, 8.9, 8.9, 8.9], 
             [8.4, 0, 0, 0, 9.2],
             [7.2, 0, 0, 0, 9.4],
             [6.1, 6.8, 7.7, 8.7, 6.1]])
ukp1 = u.copy()

# Iterate Jacobi until Convergence.
while error>tolerance:
    # Update Iteration Counter.
    k += 1
    
    # Loop through Computational nodes.
    for j in np.arange(1,Mmax):
        for i in np.arange(1,Nmax):
            ukp1[j][i] = 0.25*(u[j+1][i]+u[j][i+1]+u[j-1][i]+u[j][i-1])
            a.append(ukp1[j][i].copy())
    
    # Calculate Error using L2 Norm.
    error = np.linalg.norm(ukp1-u)     # Alternatively, error = np.sqrt(((ukp1-u)**2).sum())
    a.append(error.copy())
    
    # Storing solution in a dictionary object
    solution["{0}".format(k)] = a.copy()
    
    # Update u and a.
    u = ukp1.copy()
    a = []
    
print(u) 

# Displaying solution in a table.
table5 = pd.DataFrame.from_dict(solution, orient='index', columns=x)
table5 = table5.rename_axis("k", axis="columns") 
table5 = table5.style.set_caption("Jacobi Iterations for Laplace Equation with Dirichlet Boundary Condition")
table5
```

The Jacobo Method is inefficient as it does not make use of nodes that has been updated. A better method to use is **Gauss-Seidel Method** which uses the values of nodes that have been updated during the current step.

## 4.5b Gauss Seidel (Iterative Method)
Using row-by-column order computation, nodes to west and south ($u_{i-1,j}$ and $u_{i,j-1}$) have been updated so the Gauss-Seidel method gives: 
\begin{align} 
u^{K+1}_{i,j} &= \frac {u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K}_{i+1,j} + u^{K}_{i,j+1}} {4}\tag{*}
\end{align}
We will do the same example of Laplace Equation with Dirichlet Boundary Condition using Gauss Seidel method.

Using row-by-column order i.e. starting at first row and sweeping across all columns we get system of equations for all computational nodes for first iteration i.e. $u^1_{i,j}$ but this time we use the updated values as given by above equation (*)(south-west-east-north order).
\begin{split}
u^1_{1,1} &= \frac {(6.8 + 7.2 + 0 + 0)}{4} &= 3.5\\
u^1_{2,1} &= \frac {(7.7 + 3.5 + 0 + 0)}{4} &= 2.8\\
u^1_{3,1} &= \frac {(8.7 + 2.8 + 9.4 + 0)}{4} &= 5.225\\
u^1_{1,2} &= \frac {(3.5 + 8.4 + 0 + 8.9)}{4} &= 5.2\\
u^1_{2,2} &= \frac {(2.8 + 5.2 + 0 + 8.9)}{4} &= 4.225\\
u^1_{3,2} &= \frac {(5.225 + 4.225 + 9.2 + 8.9)}{4} &= 6.8875
\end{split}

```python
# Code to solve Laplace equation with Dirichlet BC using Gauss Seidel Method.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialising Variables.
Nmax = 4                                 # Nmax = (Number of nodes-1) in X-direction.
Mmax = 3                                 # Mmax = (Number of nodes-1) in Y-direction.
tolerance = 1e-6                         # Tolerance for Convergence Criteria.
error = 1                                # Error.
k = 0                                    # Iteration Counter.
solution = {"0":np.zeros(6)}             # Initialising dictionary object to store solutions.
a = []                                   # Empty list for storing results which will be added to Solution Dictionary.
x = ["$u^k_{1,1}$", "$u^k_{2,1}$", "$u^k_{3,1}$", "$u^k_{1,2}$", "$u^k_{2,2}$", 
     "$u^k_{3,2}$", "$\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid$"]  # Column heading for Solution Table.

# Initialising u Array with Boundary Conditions and Guess values for Computational Nodes (u0).
u = np.array([[8.9, 8.9, 8.9, 8.9, 8.9], 
             [8.4, 0, 0, 0, 9.2],
             [7.2, 0, 0, 0, 9.4],
             [6.1, 6.8, 7.7, 8.7, 6.1]])
ukp1 = u.copy()

# Iterate Gauss Seidel until Convergence.
while error>tolerance:
    # Update Iteration Counter.
    k += 1
    
    # Loop through Computational nodes.
    for j in np.arange(1,Mmax):
        for i in np.arange(1,Nmax):
            ukp1[j][i] = 0.25*(u[j+1][i]+u[j][i+1]+ukp1[j-1][i]+ukp1[j][i-1])
            a.append(ukp1[j][i].copy())
    
    # Calculate Error using L2 Norm.
    error = np.linalg.norm(ukp1-u)     # Alternatively, error = np.sqrt(((ukp1-u)**2).sum())
    a.append(error.copy())
    
    # Storing solution in a dictionary object
    solution["{0}".format(k)] = a.copy()
    
    # Update u and a.
    u = ukp1.copy()
    a = []
    
print(u) 

# Displaying solution in a table.
table6 = pd.DataFrame.from_dict(solution, orient='index', columns=x)
table6 = table6.rename_axis("k", axis="columns") 
table6 = table6.style.set_caption("Gauss Seidel Iterations for Laplace Equation with Dirichlet Boundary Condition")
table6
```

Gauss Seidel Method gives faster convergence than Jacobi method. For the example above, the convergence happened in 18 iterations as compared to 33 for Jacobi Method.

## 4.5c The Successive Over Relaxation Method(SOR)
Writing $u^{k+1}_{i,j} = u^{k}_{i,j} + R^{k}_{i,j}$ for the Jacobi Iteration formula where $R^{k}_{i,j} = $ residual (difference between two successive estimates), so that at each iteration value of $u_{i,j}$ is corrected by an amount equal to residual. SOR can improve on the convergence of iterative method by updating values using a weighted sum of the current values and the updated values. Suppose we want to acclerate convergence by adding a bigger fraction of the residual at each iteration, then we take

\begin{align}
u^{k+1}_{i,j} = u^{k}_{i,j} + \omega R^{k}_{i,j} \tag{SOR Method}
\end{align} 

where $\omega$ is called **Relaxation parameter**.
The convergence behaviour of an iterative method can be either oscillatory or monotonic. The value of $\omega$ can alter the convergence behaviour. $\omega$ is also called tuning parameter as we can specify its value as we want. $\omega$ lies in the range$[0,2]$. For $\omega < 1$, the method is called Under relaxation and for $\omega > 1$ the method is called Over relaxation.
> * **$0<\omega<1$**: This is used when the method oscillates about the solution. The difference between successive iterations is reduced (Under Relaxation).
> * **$1<\omega<2$**: This is used when the method converges monotonically towards the solution. The difference between successive iterations is increased (Over Relaxation).
> * **$\omega=1$**: This is either Jacobi or Gauss-Seidel Method.

The optimum value of $\omega$ depends on problem and often found using Trial and error.

### Point SOR Method
Consider the Gauss-Seidel method for Laplace equation above
\begin{align} 
u^{K+1}_{i,j} &= \frac {u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K}_{i+1,j} + u^{K}_{i,j+1}} {4}
\end{align}

Now by adding and subtracting $u^K_{i,j}$, the above equation can be rewritten as:
\begin{align} 
u^{K+1}_{i,j} &= u^K_{i,j} + \frac {-4u^k_{i,j} + u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K}_{i+1,j} + u^{K}_{i,j+1}} {4}
\end{align}

where 
\begin{align} 
R_{i,j} &= \frac {-4u^k_{i,j} + u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K}_{i+1,j} + u^{K}_{i,j+1}} {4}
\end{align}

Now, writing this form of Gauss-Seidel in the SOR form $u^{K+1}_{i,j} = u^{K}_{i,j} + \omega R^{K}_{i,j}$, we obtain:

\begin{align} 
u^{K+1}_{i,j} &= (1-\omega)u^K_{i,j} + \omega \bigg( \frac{u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K}_{i+1,j} + u^{K}_{i,j+1}} {4} \bigg)
\end{align}

The $2^{nd}$ term in the brackets is simply Gauss-Seidel method so it can be written as:

\begin{align} 
u^{K+1}_{i,j} &= (1-\omega) u^K_{i,j} + \omega \bar{u} ^{K+1}_{i,j}
\end{align}

where $\bar{u} ^{K+1}_{i,j}$ is the updated values obtained from using the Jacobi or Gauss-Seidel Method. This is called **Point SOR** because the value of each point is updated.

```python
# Code to solve Laplace equation with Dirichlet BC using Gauss Seidel Method + Point SOR.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Initialising Variables.
Nmax = 4                                 # Nmax = (Number of nodes-1) in X-direction.
Mmax = 3                                 # Mmax = (Number of nodes-1) in Y-direction.
tolerance = 1e-6                         # Tolerance for Convergence Criteria.
error = 1                                # Error.
k = 0                                    # Iteration Counter.
omega = 1.0975                           # Relaxation Parameter for Point SOR.
solution = {"0":np.zeros(6)}             # Initialising dictionary object to store solutions.
a = []                                   # Empty list for storing results which will be added to Solution Dictionary.
x = ["$u^k_{1,1}$", "$u^k_{2,1}$", "$u^k_{3,1}$", "$u^k_{1,2}$", "$u^k_{2,2}$", 
     "$u^k_{3,2}$", "$\mid \mathbf{u}^{K+1} - \mathbf{u}^{K} \mid$"]  # Column heading for Solution Table.

# Initialising u Array with Boundary Conditions and Guess values for Computational Nodes (u0).
u = np.array([[8.9, 8.9, 8.9, 8.9, 8.9], 
             [8.4, 0, 0, 0, 9.2],
             [7.2, 0, 0, 0, 9.4],
             [6.1, 6.8, 7.7, 8.7, 6.1]])
ukp1 = u.copy()

# Iterate Gauss Seidel until Convergence.
while error>tolerance:
    # Update Iteration Counter.
    k += 1
    
    # Loop through Computational nodes.
    for j in np.arange(1,Mmax):
        for i in np.arange(1,Nmax):
            ukp1[j][i] = 0.25*(u[j+1][i]+u[j][i+1]+ukp1[j-1][i]+ukp1[j][i-1])
            
            # Point SOR implementation
            ukp1[j][i] = (1-omega)*u[j][i] + omega*ukp1[j][i]
            
            a.append(ukp1[j][i].copy())
    
    # Calculate Error using L2 Norm.
    error = np.linalg.norm(ukp1-u)     # Alternatively, error = np.sqrt(((ukp1-u)**2).sum())
    a.append(error.copy())
    
    # Storing solution in a dictionary object
    solution["{0}".format(k)] = a.copy()
    
    # Update u and a.
    u = ukp1.copy()
    a = []
    
print(u) 

# Displaying solution in a table.
table7 = pd.DataFrame.from_dict(solution, orient='index', columns=x)
table7 = table7.rename_axis("k", axis="columns") 
table7 = table7.style.set_caption("Gauss Seidel + Point SOR for Laplace Equation with Dirichlet Boundary Condition")
table7
```

### Line SOR
We can improve Point SOR by solving for whole rows or columns of points at the same time. The Point SOR Method is written as a **Tridiagonal Linear System** where the nodes on the current row are the unknowns to be solved and the known values on the other rows form the RHS vector.
- If $N_X > N_Y$ then we solve for row j and form tridiagonal system for values of i.
- If $N_X < N_Y$ then we solve for columns i and form tridiagonal system for values of j.

Tridiagonal system can easily be solved using **Thomas Algorithm**.

pic

If we move up row-by-row the node to the south of $u_{i,j}$ will have already been updated to $u^{K+1}_{i,j-1}$ and the nodes to the east and west will be the unknown variables in the tridiagonal system. The node to the north hasn't been updated yet, thus the SOR method can be written as:

\begin{align} 
u^{K+1}_{i,j} &= (1-\omega)u^K_{i,j} + \omega \bigg( \frac{u^{K+1}_{i,j-1} + u^{K+1}_{i-1,j} + u^{K+1}_{i+1,j} + u^{K}_{i,j+1}} {4} \bigg)
\end{align}

Now, rearranging the above equation so that the nodes of current row are on LHS gives:

\begin{align} 
-\omega u^{K+1}_{i-1,j} + 4 u^{K+1}_{i,j} - \omega u^{K+1}_{i+1,j} &= 4(1-\omega)u^K_{i,j} + \omega u^{K+1}_{i,j-1} + \omega u^{K}_{i,j+1}
\end{align}

Solving this tridiagonal system will give the updated values for all nodes on row j. Writing out **line SOR Method**(letting $N=N_X$ for brevity) we get:

\begin{align} 
-\omega u^{K+1}_{0,j} + 4 u^{K+1}_{1,j} - \omega u^{K+1}_{2,j} &= 4(1-\omega)u^K_{1,j} + \omega u^{K+1}_{1,j-1} + \omega u^{K}_{1,j+1}\\
\vdots \qquad \vdots \qquad \vdots \qquad  \vdots &= \vdots \qquad \vdots \qquad \vdots \qquad  \vdots\\
-\omega u^{K+1}_{N-2,j} + 4 u^{K+1}_{N-1,j} - \omega u^{K+1}_{N,j} &= 4(1-\omega)u^K_{N-1,j} + \omega u^{K+1}_{N-1,j-1} + \omega u^{K}_{N-1,j+1}
\end{align}

If we are using Dirichlet BC, then $u^{K+1}_{0,j} = u^0_{0,j}$ and $u^{K+1}_{N,j} = u^0_{N,j}$ are known so in the first and last equation we can put these known terms on RHS. Writing the line SOR in matrix form:

\begin{gather}
\begin{pmatrix} 
4 & -\omega \\ -\omega & 4 & -\omega \\  & -\omega & 4 & -\omega \\ & & \ddots & \ddots & \ddots \\ &  &  & -\omega & 4 & -\omega \\  &  &  &  & -\omega & 4 
\end{pmatrix}
\begin{pmatrix} 
u^{K+1}_{1,j} \\ u^{K+1}_{2,j} \\ \vdots \\ u^{K+1}_{N-2,j} \\ u^{K+1}_{N-1,j}
\end{pmatrix}
=
\begin{pmatrix} 
d_1 + \omega u^{K+1}_{0,j} \\ d_2 \\ d_3 \\ \vdots \\ d_{N-2} \\ d_{N-1} + \omega u^{K+1}_{N,j} 
\end{pmatrix}
\end{gather}

where $d_i = 4(1-\omega)u^K_{i,j} + \omega u^{K+1}_{i,j-1} + \omega u^{K}_{i,j+1}$.


### Thomas Algorithm
Given a tridiagonal system of form $A\overrightarrow{x} = \overrightarrow{d}$, let $a_i, b_i and c_i$ be the lower, main and upper diagonals of $A$ and $d_i$ be the RHS values. The steps of algorithm are as follows:
> * First we perform a forward sweep modifying the $c_i$ and $d_i$ values.
$$c^{\prime}_i = 
\begin{cases} 
      \frac{c_i}{b_i} & i=1 \\
      \frac{c_i}{b_i - a_i c^{\prime}_{i-1}} & i = 2,\cdots,N-1
\end{cases}
$$
$$d^{\prime}_i = 
\begin{cases} 
      \frac{d_i}{b_i} & i=1 \\
      \frac{d_i - a_i d^{\prime}_{i-1}}{b_i - a_i c^{\prime}_{i-1}} & i = 2,\cdots,N
\end{cases}
$$
> * The solution is then obtained by Back Substitution.
\begin{align} 
      x_N &= d^{\prime}_N \\
      x_i &= d^{\prime}_{i} - c^{\prime}_{i} x_{i+1} & i = N-1,\cdots,1
\end{align}

### Optimal Value of $\omega$
For a tridiagonal system of form $A\overrightarrow{x} = \overrightarrow{d}$, the optimal value of $\omega$ is:
\begin{align} 
      \omega_{opt} = \frac{2}{1+\sqrt{1-\rho(T_J)^2}}
\end{align}
where $T_J$ is iteration matrix for Jacobi Method given by $T_J = D^{-1}(L+U)$ and $D$ are the diagonal elements of $A$ and $L$ and $U$ are Lower and Upper triangular elements of $A$ respectively with their sign reversed(i.e. $A = D - L - U$).
$\rho(T_J)$ is the largest absolute eigen value of $T_J$(known as the spectral radius of $T_J$). Since $T_J$ depends on $\omega$, we have an implicit relationship so we will need to iterate to find $\omega_{opt}$.
